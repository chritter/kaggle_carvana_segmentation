```python
import matplotlib.pyplot as plt
import numpy as np
```

Cyclical Learning Rate Implementation

```python
epochs = np.array(range(1,26)) #25 epochs
# cycle_epochs_decay=2
decay_factor = 0.5
init_lr = 1e-4
num_epochs_per_cycle = 5 
```

```python
lr = lambda x: init_lr*0.5**(x%num_epochs_per_cycle/decay_factor)
plt.plot(epochs,lr(epochs),marker='o')
plt.xlabel('epoch'); plt.ylabel('Learning rate')
```

Minimum lr is 0 and maximum learning rate is the init_lr.
